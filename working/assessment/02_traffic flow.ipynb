```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
##problem parameters
nx = 51  # try changing this number from 51 to 81 and Run All ... what happens?
dx = 11/(nx-1)
nt = 100    # for t = 6 min
dt = .06    # in min
L = 11.0
vmax = 80.0/60
rhomax = 250.0
x = numpy.linspace(0,L,nx)
```

```python
##initial conditions
rho0 = numpy.ones(nx)*10
rho0[10:20] = 50
pyplot.plot(x, rho0, color='#003366', ls='--', lw=3)
pyplot.ylim(5,55);
```

```python
#print (numpy.array(range(1, nt)))
#print (x)
```

```python
rho = rho0.copy()
t = 0.
for n in range(1, nt):  
    t = t+dt
    rhon = rho.copy() 
    rho[1:] = rhon[1:]-vmax*(1-2.*rhon[1:]/rhomax)*dt/dx*(rhon[1:]-rhon[0:-1]) 
    rho[0] = 10.0  ## boundary condition
```

```python
pyplot.plot(x, rho, color='#003366', ls='--', lw=3)
#pyplot.ylim(10,255);
```

```python
print (t)
v0 = vmax*(1-rho0/rhomax)*(1000/60)
print (min(v0))
v = vmax*(1-rho/rhomax)*(1000/60)
print (min(v))
```

```python
print (t)
v = vmax*(1-rho/rhomax)*(1000/60)
print (min(v))
print (numpy.mean(v))
```
